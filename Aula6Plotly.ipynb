```python
import plotly.express as px
import numpy as np
t =np.linspace(0,2*np.pi,100)
fig= px.line(x=t, y=np.cos(t),labels={'x': 't', 'y':'cos(t)'})
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
```

```python
np.random.seed(1)
N = 100
random_x = np.linspace(0,1,N)
random_y0 = np.random.random((N))+5
random_y1 = np.random.random((N))
random_y2 = np.random.random((N))-5
fig = go.Figure()
fig.add_trace(go.Scatter(x=random_x,y=random_y0, mode= 'markers',
                         name='markers'))
fig.add_trace(go.Scatter(x=random_x,y=random_y1, mode= 'lines+markers',
                         name='line+markers'))
fig.add_trace(go.Scatter(x=random_x,y=random_y2, mode= 'lines',
                         name='line'))
fig.write_html('grafico plotly.html')
fig.show()
```
